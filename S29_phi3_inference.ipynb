```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch
tr_model_id = "microsoft/Phi-3.5-Mini-Instruct"
peft_model_id = "phi-3.5-mini-oasst1-finetunned"
model = AutoModelForCausalLM.from_pretrained(tr_model_id, trust_remote_code=True, torch_dtype=torch.bfloat16)
tokenizer = AutoTokenizer.from_pretrained(tr_model_id)
model = PeftModel.from_pretrained(model, peft_model_id)
model = model.merge_and_unload()
```

```python
# Run text generation pipeline
prompt = "What is a large language model?"
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=1000, device=0)
result = pipe(f"[INST] {prompt} [/INST]")
print(result[0]['generated_text'])
```
